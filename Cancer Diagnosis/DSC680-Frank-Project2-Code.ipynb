```python
#modules
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
# load data into dataframe
cancerData = pandas.read_csv("Cancer_Data.csv")
```

```python
# Confirm data load
cancerData.head(10)
```

```python
# get the features I'm interested in
meanData = cancerData [["diagnosis", "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "compactness_mean", "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean"]]
```

```python
# plot radius and fractal dimension by diagnosis
M = meanData[meanData.diagnosis == "M"] #Diagnosis transfers all values of M to M data
B = meanData[meanData.diagnosis == "B"] #Diagnosis transfers all values of B to B data

plt.scatter(M.radius_mean,M.fractal_dimension_mean, color = "red", label = "Malignant", alpha = 0.3)
plt.scatter(B.radius_mean,B.fractal_dimension_mean, color = "green", label = "Benign", alpha = 0.3)

plt.xlabel("radius_mean")
plt.ylabel("fractal_dimension_mean")

plt.legend()
plt.show()
```
